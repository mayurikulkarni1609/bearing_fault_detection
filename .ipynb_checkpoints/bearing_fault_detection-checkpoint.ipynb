## Split into ~238 samples/segment for each file

```python
import numpy as np
import scipy.io

# File names and their labels
files_labels = [
    ("Time_Normal_1_098.mat", 0),
    ("IR007_1_110.mat", 1),
    ("IR014_1_175.mat", 1),
    ("IR021_1_214.mat", 1),
    ("OR007_6_1_136.mat", 2),
    ("OR014_6_1_202.mat", 2),
    ("OR021_6_1_239.mat", 2),
    ("B007_1_123.mat", 3),
    ("B014_1_190.mat", 3),
    ("B021_1_227.mat", 3),
]

window_size = 2048
X = []  # will hold all segments
y = []  # will hold all labels

for filename, label in files_labels:
    mat = scipy.io.loadmat(filename)
    
    # Get the DE signal key (it varies per file)
    de_key = [k for k in mat.keys() if 'DE_time' in k][0]
    signal = mat[de_key].flatten()
    
    # Cut into windows
    num_windows = len(signal) // window_size
    for i in range(num_windows):
        segment = signal[i*window_size : (i+1)*window_size]
        X.append(segment)
        y.append(label)

X = np.array(X)
y = np.array(y)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

## Balance Check cell

```python
unique, counts = np.unique(y, return_counts=True)
class_names = {0: 'Normal', 1: 'Inner Race', 2: 'Outer Race', 3: 'Ball'}

for u, c in zip(unique, counts):
    print(f"{class_names[u]}: {c} segments")
```

## Balance number of segments for each file to create appropriate dataset

```python
from sklearn.utils import resample

# Separate by class
X_normal = X[y == 0]
X_inner = X[y == 1]
X_outer = X[y == 2]
X_ball = X[y == 3]

# Undersample fault classes to match normal (236)
n_samples = len(X_normal)

X_inner_down = resample(X_inner, n_samples=n_samples, random_state=42)
X_outer_down = resample(X_outer, n_samples=n_samples, random_state=42)
X_ball_down = resample(X_ball, n_samples=n_samples, random_state=42)

# Combine
X_balanced = np.vstack([X_normal, X_inner_down, X_outer_down, X_ball_down])
y_balanced = np.array([0]*n_samples + [1]*n_samples + [2]*n_samples + [3]*n_samples)

print("X_balanced shape:", X_balanced.shape)
print("y_balanced shape:", y_balanced.shape)
```

## Time domain feature extraction

```python
from scipy.stats import kurtosis, skew

def extract_time_features(segment):
    mean = np.mean(segment)
    std = np.std(segment)
    rms = np.sqrt(np.mean(segment**2))
    kurt = kurtosis(segment)
    skewness = skew(segment)
    peak = np.max(np.abs(segment))
    crest_factor = peak / rms
    
    return [mean, std, rms, kurt, skewness, peak, crest_factor]

# Apply to all segments
X_time = np.array([extract_time_features(seg) for seg in X_balanced])

print("Time domain features shape:", X_time.shape)
import pandas as pd

feature_names = ['Mean', 'Std', 'RMS', 'Kurtosis', 'Skewness', 'Peak', 'Crest_Factor']
df_time = pd.DataFrame(X_time, columns=feature_names)
df_time['Label'] = y_balanced

print(df_time.head(10))
# Get first 10 ball fault segments (label 3)
df_ball = df_time[df_time['Label'] == 3].head(10)
print(df_ball)
```

## Frequency feature extraction (FFT)

```python
from scipy.stats import kurtosis, skew

def extract_frequency_features(segment, fs=48000):
    # Compute FFT
    fft_vals = np.abs(np.fft.rfft(segment))
    freqs = np.fft.rfftfreq(len(segment), d=1/fs)
    
    spectral_mean = np.mean(fft_vals)
    spectral_std = np.std(fft_vals)
    spectral_skew = skew(fft_vals)
    spectral_kurt = kurtosis(fft_vals)
    peak_freq = freqs[np.argmax(fft_vals)]
    spectral_energy = np.sum(fft_vals**2)
    
    return [spectral_mean, spectral_std, spectral_skew, 
            spectral_kurt, peak_freq, spectral_energy]

# Apply to all segments
X_freq = np.array([extract_frequency_features(seg) for seg in X_balanced])

print("Frequency domain features shape:", X_freq.shape)
```

```python
freq_feature_names = ['Spectral_Mean', 'Spectral_Std', 'Spectral_Skew', 
                      'Spectral_Kurt', 'Peak_Freq', 'Spectral_Energy']

df_freq = pd.DataFrame(X_freq, columns=freq_feature_names)
df_freq['Label'] = y_balanced

# Show Normal
print("NORMAL:")
print(df_freq[df_freq['Label'] == 0].head(5))

# Show Ball
print("\nBALL FAULT:")
print(df_freq[df_freq['Label'] == 3].head(5))
```

## WPT(wavelet packet transform) feature extraction (for ball spin fault detection)

```python
def extract_wpt_features(segment, wavelet='db4', level=4):
    wp = pywt.WaveletPacket(data=segment, wavelet=wavelet, maxlevel=level)
    
    # Get all nodes at level 4
    nodes = [node.path for node in wp.get_level(level, 'freq')]
    
    # Extract energy from each sub-band
    energies = []
    for node in nodes:
        coeffs = wp[node].data
        energy = np.sum(coeffs**2)
        energies.append(energy)
    
    return energies

# Apply to all segments
X_wpt = np.array([extract_wpt_features(seg) for seg in X_balanced])

print("WPT features shape:", X_wpt.shape)
```

## Feature combination

```python
X_final = np.hstack([X_time, X_freq, X_wpt])
print("Final feature matrix shape:", X_final.shape)
```
